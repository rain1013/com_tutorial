```python
import pandas as pd
import numpy as np 
from datetime import datetime
```

## 1. 如何擷取使用架上TABLE、建立新的暫時TABLE、欄位轉換、欄位重新命名、日期格式轉換 

```python
# 讀檔
df1 = pd.read_csv('./SOURCE_TABLE1.csv', engine='c', encoding='ANSI', low_memory=False)
```

```python
df1
```

```python
"""
PROC SQL;
CREATE TABLE TF_TXN  as 
SELECT   CUST_ID
        ,ORIGIN_TXN_DAY 
        ,INVEST_FUND_NO
        ,CASE     WHEN  ACC_CODE IN ('01','03','05','07','08') THEN AMT_NT
             WHEN  ACC_CODE IN ('02','04','06','09','10') THEN AMT_FOR*CURRENCY_RATE
            END AS  AMT
       ,put(datepart(ORIGIN_TXN_DAY),YYMMDDN8. )  as  TXN_DAY
       ,Datepart(EXT_DATE)      as  EXT_DATE    format=date9.  
       ,put(datepart(EXT_DATE) ,YYMMN6. )   as  EXT_YYMM
FROM   CMDM.TF_TFTFSS_D_TXN      
WHERE Datepart(ORIGIN_TXN_DAY) >='01OCT2020'd
AND      Datepart(ORIGIN_TXN_DAY) <='10APR2023'd       
AND      TFS_TYPE in ('1')     
;QUIT;

"""
```

## 資料轉換

```python
df1['AMT'] = df1.apply(lambda x: x['AMT_NT'] if x['ACC_CODE'] in ['01','03','05','07','08'] else x['AMT_FOR'] * x['CURRENCY_RATE'], axis=1)
# SAS date -> datetime format
df1['TXN_DAY'] = df1['ORIGIN_TXN_DAY'].apply(lambda x: datetime.strptime(x, '%d%b%Y:%H:%M:%S').strftime('%Y%m%d'))
# SAS date -> yyyymm
df1['EXT_YYMM'] = df1['EXT_DATE'].apply(lambda x: datetime.strptime(x, '%d%b%Y:%H:%M:%S').strftime('%Y%m'))
# SAS date datepart
df1['EXT_DATE'] = df1['EXT_DATE'].apply(lambda x: datetime.strptime(x, '%d%b%Y:%H:%M:%S').strftime('%d%b%Y'))
```

```python
df1
```

## 選擇

```python
df2 = df1[['CUST_ID', 'ORIGIN_TXN_DAY', 'INVEST_FUND_NO', 'AMT', 'TXN_DAY', 'EXT_DATE', 'EXT_YYMM']]\
          [(df1['TXN_DAY'] >= '20201001') & \
           (df1['TXN_DAY'] <= '20230401') & \
           (df1['TFS_TYPE'] == 1)]
```

## 2. 建立新的TABLE、排序(ORIGIN_TXN_DAY降序，AMT升序)、排序後依同ID歸戶後給序號

## 排序

```python
"""
DATA  work.TF_TXN_2 ;          
SET     work.TF_TXN ;              
BY      CUST_ID  ORIGIN_TXN_DAY  DESCENDING  AMT ;
RETAIN  STEP 0;                      
STEP=STEP+1;                  
IF FIRST.CUST_ID  THEN STEP=1;    
keep CUST_ID
    ORIGIN_TXN_DAY 
    TXN_DAY
    AMT
    STEP;
RUN;

"""
```

```python
df2.sort_values(['CUST_ID', 'ORIGIN_TXN_DAY', 'AMT'], ascending=[True, False, True], inplace=True)
```

```python
# cumcount 從0 開始
df2['STEP'] = df2.groupby(['CUST_ID']).cumcount()
```

## 3. 三個TABLE互串、TABLE間串的KEY值不同 

```python
"""
SELECT   a.CUST_ID   
        ,a.AMT
        ,a.ORIGIN_TXN_DAY 
        ,a.TXN_DAY
        ,b.TFTFM_FIRST_TXN_DAY
        ,case        when substr(a.INVEST_FUND_NO,6,1) ='0' then datepart(a.ORIGIN_TXN_DAY) 
                when substr(a.INVEST_FUND_NO,6,1) ='5' 
                and  b.TFTFM_FIRST_TXN_DAY  is not null 
                and  a.ORIGIN_TXN_DAY >=b.TFTFM_FIRST_TXN_DAY
                then datepart(b.TFTFM_FIRST_TXN_DAY)   end  as  FIRST_TXN_DAY   format=date9.
        ,c.TFTFU_CHINESE_NAME
        FROM      (
        SELECT    *
        FROM      work.TF_TXN 
        WHERE    Datepart(ORIGIN_TXN_DAY) >='01NOV2022'd  
        AND         Datepart(ORIGIN_TXN_DAY) <='31DEC2022'd  
        )   a
        LEFT JOIN (
        SELECT    DISTINCT TFTFM_TFC_ACC_ID_NO
                ,TFTFM_INVEST_FUND_NO
                ,substr(TFTFM_INVEST_FUND_NO,1,4)   as  FUND_CODE
                ,TFTFM_TFG_TRUST_CODE
                ,TFTFM_FIRST_TXN_DAY
        FROM      CMDM_UR.TF_TFTFM_D_MF
        where      Datepart(TFTFM_EXT_DATE)  between  '01NOV2022'd and '31DEC2022'd 
        )   b
        ON  a.ACC_ID_NO =b.TFTFM_TFC_ACC_ID_NO
        AND a.INVEST_FUND_NO =b.TFTFM_INVEST_FUND_NO
        AND a.ACC_CODE =b.TFTFM_TFG_TRUST_CODE

        LEFT  JOIN   (
        Select      Distinct TFTFU_FUND_ID_CODE                            
                ,substr(TFTFU_FUND_ID_CODE,1,4)   as  FUND_CODE
                ,TFTFU_CHINESE_NAME
        From        DW.tf_tftfu_m_mf  
        Where     TFTFU_EXT_DATE =  '01NOV2022'd 
        )   c
        ON b.FUND_CODE = c.FUND_CODE
        ;QUIT;

"""
```

```python
df_c = pd.read_csv('./SOURCE_TABLE3.csv', engine='c', encoding='ANSI')
```

```python
df_c['FUND_CODE'] = df_c['TFTFU_FUND_ID_CODE'].str[:4]
df_c = df_c.loc[: , ['TFTFU_FUND_ID_CODE', 'FUND_CODE', 'TFTFU_CHINESE_NAME']]
df_c.drop_duplicates(inplace=True)
```

```python
chunksize = 1e6
empty_df = pd.DataFrame()
for chunk in pd.read_csv('./SOURCE_TABLE2.csv', encoding='ANSI', chunksize=chunksize, low_memory=False):
    empty_df = pd.concat([empty_df, df1.merge(chunk[['TFTFM_TFC_ACC_ID_NO', 'TFTFM_INVEST_FUND_NO', 'TFTFM_TFG_TRUST_CODE', 'TFTFM_FIRST_TXN_DAY']].drop_duplicates(),
                                              how='left', 
                                              left_on=['ACC_ID_NO','INVEST_FUND_NO', 'ACC_CODE'], 
                                              right_on=['TFTFM_TFC_ACC_ID_NO', 'TFTFM_INVEST_FUND_NO', 'TFTFM_TFG_TRUST_CODE'])
                                              ])
```

```python
empty_df
```

```python
# 
empty_df['FUND_CODE'] =  empty_df['TFTFM_INVEST_FUND_NO'].str[:4]
final_df = empty_df.merge(df_c, how='left', left_on='FUND_CODE', right_on='FUND_CODE')

def get_firsttxnday(df):
    if df['INVEST_FUND_NO'][5] == '0':
        return df['ORIGIN_TXN_DAY']
    elif df['INVEST_FUND_NO'][5] == '5' and (df['TFTFM_FIRST_TXN_DAY'] is not np.nan) \
                                        and (df['ORIGIN_TXN_DAY'] is not np.nan)\
                                        and datetime.strptime(df['ORIGIN_TXN_DAY'], '%d%b%Y:%H:%M:%S').strftime('%Y%m%d') >= datetime.strptime(df['TFTFM_FIRST_TXN_DAY'], '%d%b%Y:%H:%M:%S').strftime('%Y%m%d'):
        return df['TFTFM_FIRST_TXN_DAY']
        
final_df['FIRST_TXN_DAY'] = final_df.apply(lambda x: get_firsttxnday(x), axis=1)
```
